# Week 3 Documentation

## Identify the Components

### 1. Inputs
The inputs are three quiz marks entered by the user:
- quiz1
- quiz2
- quiz3

### 2. Process
The program calculates the average mark using this formula:

average = (quiz1 + quiz2 + quiz3) / 3

Then the program checks if the student passes or fails.

If average >= 50, the student passes. Otherwise, the student fails.

### 3. Output
The outputs are:
- Average mark
- Pass or Fail result

## Pseudocode

START

REPEAT

Ask the user to enter quiz 1 mark

Ask the user to enter quiz 2 mark

Ask the user to enter quiz 3 mark

Calculate average = (quiz1 + quiz2 + quiz3) / 3

Display average mark

IF average >= 50 THEN

Display Pass

ELSE

Display Fail

END IF

Ask the user if they want to enter another student's marks

UNTIL answer is not yes

END

## Algorithm Structure

### Sequence
The sequence is:
1. Input quiz marks
2. Calculate average
3. Display the result

### Selection
The selection is used to decide whether the student passes or fails.

If average >= 50, the student passes.
Otherwise, the student fails.

### Iteration
The iteration is used to allow the user to enter another student's marks again.
